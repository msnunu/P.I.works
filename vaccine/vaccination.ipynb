```python
import pandas as pd
```

```python
df = pd.read_csv("country_vaccination_stats.csv")
```

```python
df
```

```python
#fill with min, if not zero
df['daily_vaccinations'] = df.groupby('country')['daily_vaccinations'].transform(lambda x: x.fillna(x.min() if x.min() is not None else 0))
```

```python
df
```

# ..........................................................................................................

```python
df.describe()
```

```python
median_per_country = df.groupby('country')['daily_vaccinations'].median()
```

```python
top_3_countries = median_per_country.nlargest(3)
```

```python
print("Top 3 country with heighest median:")
print(top_3_countries)
```
